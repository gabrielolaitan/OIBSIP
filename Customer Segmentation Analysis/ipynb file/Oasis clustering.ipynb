```python
from sklearn.preprocessing import StandardScaler

# Select features for scaling
features_scale = ['Income', 'MntTotal', 'NumWebPurchases', 'NumStorePurchases']

# Initialize the scaler
scaler = StandardScaler()

# Scale the selected features
scaled_features = scaler.fit_transform(clas[features_scale])

# Create a new DataFrame with the scaled features
scaled_clas = pd.DataFrame(scaled_features, columns=features_scale)

# Display the first few rows of the scaled DataFrame
scaled_clas.head(20)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Determine the optimal number of clusters using the elbow method
distortions = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(scaled_clas)
    distortions.append(kmeans.inertia_)
```

```python
#### Plot the elbow curve
plt.figure(figsize=(15, 5))
plt.plot(K, distortions, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
# Based on the elbow curve, let's use 4 clusters
kmeans = KMeans(n_clusters=4)
scaled_clas['Cluster'] = kmeans.fit_predict(scaled_clas)
labels = kmeans.labels_
labels
```

```python
# Calculate cluster centers
cluster_centers = pd.DataFrame(
    scaler.inverse_transform(kmeans.cluster_centers_),
    columns=features_scale
)
# Display cluster sizes and centers
print("Cluster Sizes:")
print(scaled_clas['Cluster'].value_counts())
print("\
Cluster Centers (Original Scale):")
print(cluster_centers)
```

```python
import seaborn as sns

# Create a scatter plot of Income vs Total Spending, colored by cluster
plt.figure(figsize=(14, 6))
sns.scatterplot(data=scaled_clas, x='Income', y='MntTotal', hue='Cluster', palette='deep')
plt.title('Customer Segments: Income vs Total Spending')
plt.show()

# Brief cluster descriptions
cluster_descriptions = {
    0: "Premium (High income, high spending)",
    1: "Budget Conscious (Low income, low spending)",
    2: "High-Value (Above average income and spending)",
    3: "Mid-Market (Average income and low spending)"
}

for cluster, description in cluster_descriptions.items():
    print(f"Cluster {cluster}: {description}")
```

```python
from sklearn.metrics import silhouette_score
silhouette = silhouette_score(scaled_clas, labels)
silhouette
```

```python
# Group by cluster and calculate the mean of key features
cluster_analysis = scaled_clas.groupby('Cluster')[features_scale].mean()

# Display the analysis
cluster_analysis
```
